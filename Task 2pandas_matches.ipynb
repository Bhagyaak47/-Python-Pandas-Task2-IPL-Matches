```python
import pandas as pd
a=pd.read_csv("matches (1).csv")

df=pd.DataFrame(a)
df.head(5)
```

```python
#1)Create a bar graph for all matches yearwise data.
import matplotlib.pyplot as plt

b=a["season"]
c=b.value_counts()
plt.bar(c.index,c.values,)
plt.show()

```

```python
#2)Create pie diagram of each team wining percentage on year 2008,2012,2018

fm= df[df["season"].isin([2008,2012,2018])]
for i in [2008,2012,2018]:
    sm=fm[fm["season"]==i]
    team_wins=sm["winner"].value_counts()
    total_matches=team_wins.sum()
    wp=(team_wins/total_matches)*100
    plt.pie(wp,labels=wp.index,autopct="%1.1f%%")
    plt.show
```

```python
#6) create hist as per win by runs col

df.dropna(subset=['win_by_runs'], inplace=True)

# Plotting histogram
plt.figure(figsize=(10, 6))
plt.hist(df['win_by_runs'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Win by Runs')
plt.xlabel('Win by Runs')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
